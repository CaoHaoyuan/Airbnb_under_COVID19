## Generate total reviews per month for multiple cities

```python
df1 = pd.read_csv("Los Angeles/reviews.csv")
print(df1.shape)
df1.date = pd.to_datetime(df1.date)
df1['month'] = df1.date.dt.month
df1['year'] = df1.date.dt.year
df1
```

```python
df2 = pd.read_csv("Oakland/reviews.csv")
print(df2.shape)
df2.date = pd.to_datetime(df2.date)
df2['month'] = df2.date.dt.month
df2['year'] = df2.date.dt.year
df2
```

```python
num_reviews_per_month1 = df1.groupby(['year','month']).size().reset_index()
num_reviews_per_month1
```

```python
num_reviews_per_month2 = df2.groupby(['year','month']).size().reset_index()
num_reviews_per_month2
```

```python
num_reviews_per_month = pd.merge(num_reviews_per_month1, num_reviews_per_month2, on=['year','month'])
num_reviews_per_month["0_x"] = num_reviews_per_month["0_x"] + num_reviews_per_month["0_y"]
num_reviews_per_month
```

```python
num_reviews_per_month = num_reviews_per_month.drop('0_y', axis = 1)
num_reviews_per_month
```
